```python
# one_hot_encoded_columns = pd.get_dummies(X, columns = ['allocation', 'status'])
encoder = OneHotEncoder(sparse_output=False)
encoded_data = encoder.fit_transform(X[['allocation','status']])
# encoded_status = encoder.fit_transform(X[['status']])

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['allocation','status']))
final_df = pd.concat([X, encoded_df], axis=1)
final_df = final_df.drop(columns = ['allocation','status'])
final_df.info()


```

```python
# final_df['province.id'].astype(float)

final_df['province_id'] = final_df['province.id'].str.extract('(\d+)').astype(float)
final_df = final_df.drop(columns=['province.id'])

```

```python
# final_df = final_df[['is_clean','is_new','is_instant','is_plus']].replace({True: 1, False: 0})
final_df['is_clean'] = final_df['is_clean'].astype(int)
final_df['is_new'] = final_df['is_new'].astype(int)
final_df['is_instant'] = final_df['is_instant'].astype(int)
final_df['is_plus'] = final_df['is_plus'].astype(int)
final_df.info()
```

```python
final_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.metrics import mean_squared_error



X_train, X_test, y_train, y_test = train_test_split(final_df, y, test_size=0.2, random_state=42)
X_train.shape[1]
```

```python
# To-Do
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
from sklearn.metrics import r2_score


model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),  
    Dense(32, activation='relu'), 
    Dense(1,activation='sigmoid') 
])




# model = RandomForestClassifier(n_estimators=100, random_state=42)
# model.fit(X_train, y_train)
# y_pred = model.predict(X_test)
# accuracy = accuracy_score(y_test, y_pred)
# print(f"Accuracy: {accuracy * 100:.2f}%")

```

```python
model.compile(optimizer='adam', loss='mae',metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=100, batch_size=60, validation_split=0.2)

```

```python
# Step 7: Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss:.2f}, Test MAE: {test_mae:.2f}")

# Step 8: Make predictions
y_pred = model.predict(X_test)
```
